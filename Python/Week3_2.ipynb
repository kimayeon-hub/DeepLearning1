# 타이타닉 데이터 분석하기
- 성별과 나이에 따른 생존율 예측하기

## 1. train dataset 살펴보기
1. train dataset을 엑셀로 열기
2. ctrl + a 눌러서 모든 데이터 선택하기
3. 데이터 > 피벗 테이블 > 실행
4. 피벗 테이블 필드 목록 > 보고서에 추가할 필드 선택 (Survived, sex)
5. 생존자들의 **성별 비율**을 보기 위해 <br>
  - Name 필드를 추가로 선택 (모든 사람의 이름이 다 있다고 가정)
  - Name을 데이터 영역으로 옮기기
  image.png
  > 여성 승객 총 314명 중 233명이 생존하였고 <br>
  > 남성 승객 총 577명 중 109명이 생존하였다는 것을 알 수 있음
6. **좌석 클래스 별 생존 비율**을 보기 위해 <br>
  - Pclass 필드를 추가로 선택하기
  - Pclass를 기준으로 생존했는지 안 했는지를 보고 싶은 것이므로 행 영역으로 옮기기
  image.png
  > 1등석 승객들이 생존확률이 높고 여성승객이 생존확률이 높았다는 것을 알 수 있음
7. 나이로 나눠서 **아이들과 어른들의 생존 비율**을 보기 위해 <br>
  - raw data에 Age 필드 뒷부분에 Children인지 아닌지를 나타내는 Children 필드 추가하기
      - Children 필드의 첫번째 데이터 부분에 `=if(F2<18, 1, 0)`를 하면 간단하게 수식을 만들 수 있음
      > F2부분의 값이 18보다 작으면 1, 아니면 0이라는 뜻
  - Children 필드의 첫번째 데이터를 클릭한 다음, 셀의 오른쪽 하단 모서리에 있는 작은 팦란색 원을 더블 클릭하면 그 열 전체에 함수가 적용 됨
  - 변경한 파일을 저장하고, 모든 데이터를 선택(ctrl + a)한 뒤 피벗 테이블 열기
  - Survived(열), Name(전체 수를 알기 위한 필드)(열), Children (행) 선택

  > 아이와 어른의 생존 비율이 크게 다르지 않다는 것을 알 수 있음

## 2. test dataset에 생존율을 예측해서 저장하기
1. Passenger column 옆에 Survived column 추가하기
2. `=if(E2="male",0,1)`으로 값 채우고 저장
> 남자면 사망, 여자면 생존이라고 예측을 한 것
3.` =IF(E2="female",1,IF(C2=1,1,0))`라고 채우면
> 여자면 생존, 남자인데 1등석이면 생존, 아니면 생존X 라고 예측한 것
4. 수식으로 들어가 있는 것이므로 값으로 변경시키기
  - Survived 열을 다 드래그(ctrl + shift + downarrow) 해서 복사하기
  - 안 쓰는 열에 다가 `우클릭 > 골라 붙여넣기 > 값`해서 붙여넣기
  - 붙여넣기 한 값을 복사해서 Survived에 넣기
  - Survived에 있는 값 클릭해보면 값으로 변경되어 있는 것을 알 수 있음
5. Passenger과 Survived를 제외하곤 필요없으므로 다 삭제하기
6. `kaggle > titanic 대회`에 test dataset 제출하기
