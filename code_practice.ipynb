# 타이타닉 데이터의 수치화 및 정규화

```python
# train.csv 파일 읽어오기
import pandas as pd
data_train = pd.read_csv('train.csv')

# 데이터를 numpy 배열에 저장하기 위해 0으로 된 numpy 배열 생성
data_train_np = np.zeros((data_train.shape[0], 8))

# 성별(Sex) 데이터를 원 핫 인코딩하기
from tensorflow.keras.utils import to_categorical
sex = np.zeros((data_train.shape[0]))
sex[data_train['Sex'] == 'female'] = 1
data_train_np[:, :2] = to_categorical(sex)

# 객실 등급(Pclass) 데이터를 원 핫 인코딩하기
data_train_np[:, 2:5] = to_categorical(data_train['Pclass'].to_numpy() - 1)

# 나이(Age), 형재 자매 배우자 수(SibSp), 부모 자녀 수(Parch)는 정규화하기
data_train_np[:, 5] = data_train['Age'] / 80
data_train_np[:, 6] = data_train['SibSp'] / 10
data_train_np[:, 7] = data_train['Parch'] / 10

# 나이 데이터에 있는 결측치를 평균값으로 대체하기
data_train_np[np.isnan(data_train_np)] = 30/80

data_train_np
```

```python
# test.csv 파일 읽어오기
import pandas as pd
data_test = pd.read_csv('test.csv')

# 데이터를 numpy 배열에 저장하기 위해 0으로 된 numpy 배열 생성
import numpy as np
data_test_np = np.zeros((data_test.shape[0], 8))

# 성별(Sex) 데이터를 원 핫 인코딩하기
from tensorflow.keras.utils import to_categorical
sex = np.zeros(data_test.shape[0])
sex[data_test['Sex'] == 'famale'] = 1
data_test_np[:, :2] = to_categorical(sex)

# 객실 등급(Pclass) 데이터를 원 핫 인코딩하기
data_test_np[:, 2:5] = to_categorical(data_test['Pclass'].to_numpy() - 1)

# 나이(Age), 형재 자매 배우자 수(SibSp), 부모 자녀 수(Parch)는 정규화하기
data_test_np[:, 5] = data_test['Age'] / 80
data_test_np[:, 6] = data_test['SibSp'] / 10
data_test_np[:, 7] = data_test['Parch'] / 10

# 나이 데이터에 있는 결측치를 평균값으로 대체하기
data_test_np[np.isnan(data_test_np)] = 30 / 80

data_test_np
```

```python
# 딥러닝 학습에 사용하기 위한 출력 데이터 뽑아내기
data_train_np_y = to_categorical(data_train['Survived'].to_numpy())
data_train_np_y
```

```python
# 케라스 라이브러리 가져오기
from keras.models import Sequential
from keras import layers
from keras import optimizers

# 한 방향으로 데이터가 흐르는 딥러닝의 기본 모델 구조를 구성하기
model = Sequential()

# 완전 연결층 5개 쌓기
# 노드의 수는 8개, 활성화 함수로는 relu를 사용하라
model.add(layers.Dense(units=8, activation='relu'))

# 노드의 수는 16개, relu 사용
model.add(layers.Dense(units=16, activation='relu'))

# 노드의 수는 32개, relu 사용
model.add(layers.Dense(units=32, activation='relu'))

# 노드의 수는 64개, relu 사용
model.add(layers.Dense(units=64, activation='relu'))

# 노드의 수는 2개, softmax 사용
model.add(layers.Dense(units=2, activation='softmax'))

# 컴파일하기 (손실함수, 최적화 도구 등 학습에 필요한 요소들을 설)
model.compile(optimizer=optimizers.Adam(learning_rate=0.002), loss='categorical_crossentropy', metrics=['accuracy'])

# 모델 학습시키기
# 데이터 전체를 300번씩 학습하고, 100개의 데이터를 학습했을 때마다 가중치 업데이트하고, validation dataset으로는 30% 사용하기
hist = model.fit(data_train_np, data_train_np_y, epochs=300, batch_size=100, validation_split=0.3)
```

# 집 가격 예측하기

```python
import pandas as pd
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
```

```python
# NaN 비율 계산 함수
def check_nan_ratio(df):
  # 비율을 퍼센트로 계산하기
  return df.isna().mean() * 100

# 결측치 비율이 5% 이상인 컬럼 제거 함수
def drop_high_nan_cloumns(train_df, test_df, threshold=5.0):
  # 학습 데이터에서 결측치 비율 계산
  train_nan_ratio = check_nan_ratio(train_df)

  # 5% 이상인 컬럼 추출
  cols_to_drop = train_nan_ratio[train_nan_ratio >= threshold].index

  # 학습 데이터에서 제거
  train_df_cleaned = train_df.drop(columns=cols_to_drop)

  # 테스트 데이터에서도 동일한 컬럼 제거
  test_df_cleaned = test_df.drop(columns=cols_to_drop)

  # 전처리한 데이터와 제거된 컬럼명 반환
  return train_df_cleaned, test_df_cleaned, cols_to_drop

train_df_cleaned, test_df_cleaned, removed_columns = drop_high_nan_cloumns(data_train, data_test, 5)
```

```python
# train_df_cleaned에 NaN값이 아직 남아 있다면, 그 행 지우기
train_df_cleaned = train_df_cleaned.dropna()

# test_df_cleaned에서 NaN이 존재하는 컬럼 찾기
nan_cols = test_df_cleaned.columns[test_df_cleaned.isna().any()]

# 해당 컬럼들을 두 데이터셋에서 제거
train_df_cleaned = train_df_cleaned.drop(columns=nan_cols, errors='ignore')
test_df_cleaned = test_df_cleaned.drop(columns=nan_cols, errors='ignore')

# train_df_cleaned에서 'Id'컬럼 제거
if 'Id' in train_df_cleaned.columns:
  train_df_cleaned = train_df_cleaned.drop(columns=['Id'])

# 결과 출력
print(f"test_df_cleaned에서 NaN이 있어 제거된 컬럼들: {nan_cols}")
print(f"최종 train_df_cleaned shape: {train_df_cleaned.shape}")
print(f"최종 test_df_cleaned shape: {test_df_cleaned.shape}")
```

```python
# 모든 값이 동일한 컬럼 찾기
constant_cols = [col for col in train_df_cleaned.columns if train_df_cleaned[col].nunique() == 1]

# 해당 컬럼 제거
train_df_cleaned = train_df_cleaned.drop(columns=constant_cols)
```

```python
# test 데이터에서 Id 컬럼 제거 및 저장
if 'Id' in test_df_cleaned.columns:
  id_column = test_df_cleaned['Id']
  test_df_cleaned = test_df_cleaned.drop(columns=['Id'])

# train 데이터에서 숫자가 아닌 컬럼 찾기 (SalePrice 제외)
non_numeric_cols = train_df_cleaned.select_dtypes(exclude=[np.number]).columns
if 'SalePrice' in non_numeric_cols:
  non_numeric_cols.remove('SalePrice')

# train 데이터에서 원 핫 인코딩 후 컬럼 수 계산
total_columns = 0
encoding_info = {}  # 각 컬럼별 인코딩 정보를 저장할 딕셔너리

for col in train_df_cleaned.columns:
  if col == 'SalePrice':
    total_columns += 1
  elif col in non_numeric_cols:
    unique_values = train_df_cleaned[col].unique()
    encoding_info[col] = unique_values
    total_columns += len(unique_values)
  else:
    total_columns += 1

# 학습 데이터와 테스트 데이터를 같은 기준으로 변환하기 위해 직접 원-핫 인코딩하는 것
def one_hot_encode_value(value, unique_values):
  # 원 핫 인코딩 벡터를 반환
  one_hot_vector = np.zeros(len(unique_values))
  if value in unique_values:
    index = np.where(unique_values == value)[0][0]
    one_hot_vector[index] = 1.0
  return one_hot_vector

# 변환 함수
def transform_dataframe(df, encoding_info, is_train=True):
  current_col = 0
  num_cols = total_columns if is_train else total_columns - 1
  transformed_array = np.zeros((df.shape[0], num_cols))

  for col in train_df_cleaned.columns:
    if col == 'SalePrice' and not is_train:
      continue;
    if col in non_numeric_cols:
      unique_values = encoding_info[col]
      for i, value in enumerate(df[col]):
        one_hot_vector = one_hot_encode_value(value, unique_values)
        transformed_array[i, current_col:current_col + len(one_hot_vector)] = one_hot_vector
      current_col += len(unique_values)
    else:
      transformed_array[:, current_col] = df[col]
      current_col += 1
  return transformed_array

# 학습 데이터 변환
train_array = transform_dataframe(train_df_cleaned, encoding_info, True)
# 테스트 데이터 변환
test_array = transform_dataframe(test_df_cleaned, encoding_info, False)
```

```python
# train_array의 데이터 분리
train_x = train_array[:, :-1]
train_y = train_array[:, -1]
test_x = test_array
```

```python
# train_x의 최댓값과 최솟값 계산
train_x_min = np.min(train_x)
train_x_max = np.max(train_x)

# train_y의 최댓값과 최솟값 계싼
train_y_min = np.min(train_y)
train_y_max = np.max(train_y)

# 정규화 함수
def normalization(data, min_val, max_val):
  return (data - min_val) / (max_val - min_val)

# 정규화
train_x = normalization(train_x, train_x_min, train_x_max)
train_y = normalization(train_y, train_y_min, train_y_max)
test_x = normalization(test_x, train_x_min, train_x_max)
```
